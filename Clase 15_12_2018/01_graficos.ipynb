```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [4,3]
plt.rcParams['figure.dpi'] = 150
```

```python
xvals = np.arange(0,10,0.1)
plt.plot(xvals)
```

```python
plt.plot(xvals,np.sin(xvals))
```

```python
plt.plot(xvals,np.sin(xvals),label = r'$y = \sin(x)$')
plt.plot(xvals,np.cos(xvals),label = r'$y = \cos(x)$')
plt.plot(xvals,np.tan(xvals),label = r'$y = \tan(x)$')
plt.xlabel('$Valores en X$')
plt.ylabel('$Valores en y$')
plt.legend()
```

```python
plt.plot(xvals,np.sin(xvals),label = r'$y = \sin(x)$',color = 'y')
plt.plot(xvals,np.cos(xvals),label = r'$y = \cos(x)$',marker = 'o',color = 'b')
plt.xlabel('$Valores en X$')
plt.ylabel('$Valores en y$')
plt.legend()
```

```python
rands = np.random.normal(1000)
print(rands)
plt.xlabel('Iteration #')
plt.ylabel('Lo que sea')
plt.plot()
```

```python
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(
    np.random.normal(size = (512,512)),sigma = 10)
#print(rands2d)
plt.imshow(rands2d)
plt.colorbar()
plt.plot()
```
