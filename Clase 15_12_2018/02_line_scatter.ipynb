```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [4,3]
plt.rcParams['figure.dpi'] = 150
```

```python
xvals = np.arange(0,10,0.1)
plt.plot(xvals,np.sin(xvals),label = r'$y = \sin(x)$')
plt.plot(xvals,np.cos(xvals),label = r'$y = \cos(x)$')
plt.xlabel('$Valores en X$')
plt.ylabel('$Valores en y$')
```

```python
xvals = np.arange(0,10,0.1)
idx = 0 
for marker in ('-','--',':','-.'):
    plt.plot(xvals,np.sin(xvals)+idx,''.join(('k',marker)))
    idx += 1
```

```python
xvals2 = np.arange(0,10,0.1)
plt.plot(xvals2,np.sin(xvals2),'k.--')
```

```python
xvals2 = np.arange(0,10,0.1)
plt.plot(xvals2,np.sin(xvals2),'r.',markersize = 1, markevery = 1)
```
